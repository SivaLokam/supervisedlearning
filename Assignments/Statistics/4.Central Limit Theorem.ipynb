### Sample programme to calculate probability

```python
def calc_prov(pop_mean,
              standard_deviation,
              sample_size,
              random_value=None,
              mean_value=None,
              prob_random_variable=True,
              prob_for_mean=False):
    """
    Sample programme to 
    i) calculate probabiliy for normal random variable at a given value and
    ii) calculate probaility for at mean value of the sample of size N 
           
    """
    upop = pop_mean
    usample = pop_mean
    sd = standard_deviation
    N = sample_size
    sd_sample = standard_deviation/(np.sqrt(N))
    
    if prob_random_variable:
        zrv = (random_value - upop) / sd
        p_rv = norm.cdf(zrv)
    if prob_for_mean:
        zmv = (mean_value - usample) / sd_sample
        p_mv = norm.cdf(zmv)
    
    return p_rv,p_mv
```
