### Basic probability

<div style="background-color:lightyellow">    
<b> <font color='blue'>1. Two dies are rolled at once. Find out the probability for sum of numbers being even and one of the die shows 6. <font></b>

<b>Analysis: </b>
It is single event, the total no of possible combinations are : 36 ( as there are two dice)
Total no of favourable outcomes  = (sum is even and one of the die should show 6) possible combinations are : (6,2) (6,4),(6,6),(2,6),(4,6)  

<b>Answer:</b>
 probability = (No of favourable outcomes / Total no of possible outcomes) = (5/36) 
<hr>
<b> <font color='blue'> 2. Two dies are rolled at once. Find out the probability for sum of numbers being less than 7.</b>

<b>Analysis: </b>
No of favourable outcomes are : (1,5)(1,4),(1,3),(1,2),(1,1) ,(2,4),(2,3),(2,2),(2,1),(3,3),(3,2),(3,1),(4,2),(4,1),(5,1)

<b>Answer:</b>
probability = (No of favourable outcomes / Total no of possible outcomes)     = (15/36)

### Conditional, Joint and Marginal Probability

<div style="background-color:lightyellow">
<b><font color='blue'> 5. In my town, it's rainy for one third of the days. Given that it is rainy, there will be heavy traffic with probability 1/2, and given that it is not rainy, there will be heavy traffic with probability 1/4. If it's rainy and there is heavy traffic, I arrive late for work with probability 1/2. On the other hand, the probability of being late is 1/8 if it is not rainy and there is no heavy traffic. In other situations (rainy and no traffic, not rainy and traffic) the probability of being late is 0.25, 0.25. You pick a random day.</font><b>
</div>

<div style="background-color:lightyellow">
<b><font color='blue'> a) What is the probability that it's not raining and there is heavy traffic and I am not late? </font></b>
    
<b>Analysis: </b>

p(NotRaining ^ HeavyTraffic ^ Not Late) = p(NotLate|(HeavyTraffic & Not rainy)) X p(HeavyTraffic|not rainy) X p(not rainy)

<b>Answer: </b>
p(NotRaining ^ HeavyTraffic ^ Not Late) = (3/4) X (1/4) X (2/3) = (1/8)

<b><font color='blue'>b) What is the probability that I am late? </font></b>

<b>Analysis: </b>

p(Late) = p(Late| (HT & Rainy)) + p(Late| (NHT & Rainy) + p(Late| (HT & Not Rainy)) + p(NHT & Not Rainy)

<b>Answer: </b>
p(Late) = (1/2 X 1/2 X 1/3) + (1/4 X 1/2 X 1/3) + (1/4 X 1/4 X 2/3) + (1/8 X 3/4 X 2/3) = 11/48 

<b><font color='blue'>c) Given that I arrived late at work, what is the probability that it rained that day? </font></b>
    
<b>Analysis: </b>
p(rainy|Late) =p(Late|rainy) X p(Late) / p(rainy) 

             = [p(Late|(HT&rainy)) + p(Late|(NHT & rainy))] X p(Late) / p(rainy)

<b>Answer: </b>
p(rainy|Late) = [(1/2 X 1/2) + (1/4 X 1/2)] X 11/48 /(1/3) = 33/128

* 70% purchase a cup of coffee
* 40% purchase a piece of cake
* 20% purchase both a cup of coffee and a piece of cake. 

<b><font color='blue'>Given that a randomly chosen customer has purchased a piece of cake, what is the probability that he/she also purchased a cup of coffee.</font></b>
</div>

<div style="background-color:lightyellow">
<b>Analysis: and Answer: </b>

p(coffee) = 0.7 , p(cake) = 0.4, p(coffe and cake) = 0.2

p(coffee|cake) = p(coffe and cake) / p(cake) = 0.2 / 0.4 = 0.5

<b><font color='blue'>8. A is known to tell the truth in 5 cases out of 6 and he states that a white ball was drawn from a bag containing 8 blacks and 1 white ball. Find the probability that the white ball was drawn.</font></b>


</div>

<div style="background-color:lightyellow">
<b>Analysis: and Answer: </b>

p(Truth|white) = p(white|Truth) X p(Truth) / p(white)

p(white) = p(white|Truth) X p(Truth) + p(white|False) X p(False)

         = (1/9 X 5/6) + (8/9 X 1/6) = 13 /54


p(Truth|white) = (1/9 X 5/6) / (13/54) = 5/13
