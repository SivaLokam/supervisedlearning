### <span id="clusters" style="color:Green"> 5.2 Scatter Plot to show clusters </span> <br>
While we have seen a scatter plot of random dots before, scatter plots are most useful with representing discrete data points that show a trend or clusters. Each data series will be plotted in different color per plotting command by default, which helps us distinguish the distinct dots from each series.

```python
np.random.seed(123)
x0 = np.random.rand(10)+2
y0 = np.random.rand(10)+2.5
np.random.seed(456)
x1 = np.random.rand(10)+4
y1 = np.random.rand(10)+2.5

plt.scatter(x0,y0)
plt.scatter(x1,y1)
plt.show()
```

We can observe from above plot there are two artifically created clusters of dots colored blue(left half) and orange(right half)

There is another way to generate clusters and plot them on scatter plots. For this, we make use of make_blobs() function found in sklearn package.

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=10, centers=3, n_features=2,random_state=123)

print(X.shape)

print(y.shape)

plt.scatter(X[:,0],y)
plt.scatter(X[:,1],y)
plt.show()
```

### <span id="trendline" style="color:Green"> 5.3 Adding trendline over a scatter plot  </span>

Multiple plot types can be overlaid on top of each other. For example, we can add a trendline over a scatter plot.

```python
help(np.polyfit)
```

```python
x = list(range(10))
y = x + np.random.rand(10)*3-0.5

m,c = np.polyfit(x,y,1)
y_hat = [m*x1 +c for x1 in x ]
plt.scatter(x,y)
plt.plot(x,y_hat,'black')
plt.show()
```

#### <span id="c65" style="color:Green"> 6.5 Adding legend  </span>

Inorder to adjust the location we pass loc settings. Possible loc settings are as follows:-

<ul>
	<li>'best':0(onlyimplementedforaxeslegends)</li>
	<li>'upperright':1</li>

```python
d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]


%matplotlib inline
plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)
plt.grid(True,linewidth=0.5)
plt.grid(linestyle="--",color="#888888",alpha=1)
plt.legend(('Tokyo','Hawaii','Hong Kong'))
plt.xlabel('Date',size=12,fontweight='bold')
plt.ylabel('Temperature (C)',size=12,fontweight='semibold')
plt.title('Daily Temps of 3 cities in the second week of December',size=16,fontweight='bold')
plt.show()
```

Inorder to save a figure we put plt.savefig(outputpath) at the end of plotting commands.If you want to save the figure as a file as well as display it on the notebook output, you can call both plt.savefig() and plt.show().

<h4> 1.1 Color Cycle </h4>

A color cycle is a list of colors used to control the color of a series automatically, such as each data series in multiline plots.These colors are designed to show good contrast between distinct categories.Each color is named 'C0' to 'C9', and can be called in manually by specifying a color in the preset color cycle.

```python
i = 3
plt.plot([i]*6)
plt.xlim(0,10)
plt.grid()
plt.show()
```

```python
colors=['b','g','r','c','m','y','k','w']
plt.figure(figsize=(4,4))
for color in colors:
#     color = np.random.choice(colors)
    plt.plot([np.random.randint(10)]*5,c=color,label=color) 
plt.legend()
plt.xlim(0,6)
plt.show()
```

#### 1.2 Single-letter abbreviations for basic colors
There are several common colors with built-in. Some are listed below:-

* 'b':Blue

* 'c':Cyan

* 'y':Yellow
* 'k':Black
* 'w':White

```python
colors=['b','g','r','c','m','y','k','w']
for color in colors:
#     print(color)
    plt.plot(np.random.randint(5)*5,c=color,label=color)
plt.legend()
plt.show()
```

#### 1.3 RGB or RGBA color code
A color can be specified as a tuple of three or four float numbers between 0 to 1 such as (0.1,0.1,0.2) or (0.2,0.2,0.3,0.8).The first three colors define how red, green and blue should be mixed into the desired color output. The optional fourth value is the alpha value to control the transparency level.

#### 2.1 Histogram

Histograms are useful in surveying the distribution of data. For example, we can plot data on a histogram when we want to see some age groups distributed in a population, light exposure in a photograph, or the amount of precipitation in each month in a city.

In Matplotlib, we call the plt.hist() function with a linear array. Matplotlib will automatically group them into 'bins' and plot out the frequencies for each bin in bars

```python
np.random.binomial(10,0.6)
```

```python
# help(np.random.binomial)
```

```python
np.random.binomial(10,0.9,100)
```

```python
prob = [round(x*0.1,1) for x in list(range(5,10,1))]
```

```python
for p in prob:
    x = np.random.binomial(100,p,10000)
    plt.hist(x,bins=40)
    plt.show()
```

#### 2.2 Bar plots

Bar plots are useful for comparing absolute levels of discrete data series. They are created by the function plt.bar(labels,heights) in Matplotlib.

```python
# Let's take an example of the market capitalization of today's much hyped cryptocurrencies.

# Data series consists of top 5 cryptocurrencies in terms of market capitalization
cc =['BTC','XRP','ETH','BCH','ADA']
cap = [282034,131378,107393,49999,26137]

plt.bar(cc,cap)

plt.title('Market capitalization of five top cryptocurrencies in Jan')
plt.xlabel('Crytocurrency')
plt.ylabel('Market capitalization (million USD)')
plt.show()
```

To create bar plots with bars in the designated order, we can make use of Pandas and its Matplotlib integration. The procedure is as follows:

1. Create a Pandas DataFrame df
2. Plot the bar chart with df.plot(kind='bar')
3. Set the labels of xticks
4. Adjust the other plot properties
5. Show the plot with plt.show()

```python
df = pd.DataFrame({"CC":cc,"Cap":cap})
df
```

```python
ax = df.plot(kind='bar')
ax.set_xticklabels(df['CC'])
plt.title('Market capitalization of five top cryptocurrencies in Jan')
plt.xlabel('Crytocurrency')
plt.ylabel('Market capitalization (million USD)')
plt.show()
```

### 2.3 Box Plot

Box Plots are most suited for displaying the distribution of a variable across multiple groups. The bottom and top of the boxes indicate the lower and upper quartiles, respectively, and the line inside the box is for the median. Vertical lines extending from the boxes("whiskers") show the range of the data(this is 1.5x past the upper and lower quartiles in matplotlib).

Box Plots can be thought of as an hybrid between histograms and bar plots.They surface much of the same information as bar plots, but they also expose the variation in the data.

```python
tips = sns.load_dataset('tips') 
tips.shape
```

```python
sns.set_style('whitegrid')
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.boxplot(x=tips['day'],y=tips['tip'])
plt.show()
```

### 2.4 Pie Chart
Pie chart is a circular representation of component ratios.The angle and hence the arc length of each sector ratio(called 'wedges') presents the proportion that each component accounts for , relative to the whole. Matplotlib uses plt.pie() function to draw pie charts. We can label each sector with 'labels' as well as percentage with 'autopct' automatically.

```python
# Example of web server using in first week of January-2017
x=[0.31,0.3,0.14,0.1,0.15]
labels=['nginx','Apache','IIS','Varnish','Others']
plt.pie(x,labels=labels,autopct='%1.1f%%')
plt.title('Web Server Usage Statistics')
plt.show()
```

### 3. Drawing Subplots
When you have multiple plots it is necessary to organize them into panels in the same figure, such as when illustrating different aspects of the same dataset.Matplotlib provides a few ways to create figures with multiple subplots.

#### 3.1 Initiating a figure with plt.figure()
The plt.figure() is an API that is used to initiate a figure that serves as the base canvas. It takes in arguments that determine the number of figures and parameters such as size and background color of the plot image. It displays new area of the canvas for plotting 'axes' when called.We wouldnt obtain any graphical output unless we add other plotting elements.

If we were to call plt.show(), we would see a Matplotlib 'figure' object being returned , as shown below:

### 3.2 Initiating subplots as axes with plt.subplot()
To initiate the axes plot instances that actually from each plot,we use plt.subplot(). It takes 3 parameters: rows, columns and plot number.

```python
# Initiates a figure area for plotting
fig = plt.figure()

# Initiates six subplot axes
ax1 = plt.subplot(231)
ax2 = plt.subplot(232)
ax3 = plt.subplot(233)
ax4 = plt.subplot(234)
ax5 = plt.subplot(235)
ax6 = plt.subplot(236)

# Print the type of ax1
print(type(ax1))

# Label each subplot with corresponding identities
ax1.text(0.3,0.5,'231',fontsize=18)
ax2.text(0.3,0.5,'232',fontsize=18)
ax3.text(0.3,0.5,'233',fontsize=18)
ax4.text(0.3,0.5,'234',fontsize=18)
ax5.text(0.3,0.5,'235',fontsize=18)
ax6.text(0.3,0.5,'236',fontsize=18)

plt.show()
```

Note: Printing of the axes returns <class 'matplotlib.axes._subplots.AxesSubplot'> as a result.

```python
fig,axs = plt.subplots(nrows=2,ncols=3,sharex=True,sharey=True)
axs = axs.flatten()
for axs,ind in zip(axs,range(len(axs))):
    axs.text(0.3,0.5,'23'+str(ind+1),fontsize=18)
```
