### <span id="overlay" style="color:Green">Overlaying multiple data series in a plot </span>

#### <span id="overlay" style="color:Green"> 5.1 Multiline plots </span>

To create a multiline plot, we can draw a line plot for each data series before concluding the figure.

```python
# Let's try plotting the temperatures of three different cities with the following code
# Prepare the data series
d = [11,12,13,14,15,16,17]
a1 = [15.3,15.4,12.6,12.7,13.2,12.3,11.4]
a2 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
a3 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]

plt.plot(d,a1)
plt.plot(d,a2)
plt.plot(d,a3)
plt.show()
```

### <span id="clusters" style="color:Green"> 5.2 Scatter Plot to show clusters </span> <br>
While we have seen a scatter plot of random dots before, scatter plots are most useful with representing discrete data points that show a trend or clusters. Each data series will be plotted in different color per plotting command by default, which helps us distinguish the distinct dots from each series.

```python
np.random.seed(123)
x0 = np.random.rand(10)+2
y0 = np.random.rand(10)+2.5
np.random.seed(456)
x1 = np.random.rand(10)+4
y1 = np.random.rand(10)+2.5

plt.scatter(x0,y0)
plt.scatter(x1,y1)
plt.show()
```

We can observe from above plot there are two artifically created clusters of dots colored blue(left half) and orange(right half)

There is another way to generate clusters and plot them on scatter plots. For this, we make use of make_blobs() function found in sklearn package.

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=10, centers=3, n_features=2,random_state=123)

print(X.shape)

print(y.shape)

plt.scatter(X[:,0],y)
plt.scatter(X[:,1],y)
plt.show()
```

### <span id="trendline" style="color:Green"> 5.3 Adding trendline over a scatter plot  </span>

Multiple plot types can be overlaid on top of each other. For example, we can add a trendline over a scatter plot.

```python
help(np.polyfit)
```

```python
x = list(range(10))
y = x + np.random.rand(10)*3-0.5

m,c = np.polyfit(x,y,1)
y_hat = [m*x for x in x +c]
plt.scatter(x,y)
plt.plot(x,y_hat,'black')
plt.show()
```

### <span id="c6" style="color:Green"> 6. Adjusting axes, grids, labels, titles, and legends  </span>

Till now we learnt how to convert numeric values to dots and lines with Matplotlib. Good visualization often requires more design input to suit our custom data visualisation needs and purpose.Text labels are needed to make figures informative in many cases.

#### <span id="c61" style="color:Green"> 6.1 Adjusting axis limits  </span>

Instead of matplotlib automatically choosing the x and y axis limits. We can set the limits of x and y axes using plt.xlim() and plt.ylim() respectively.

```python
d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]


%matplotlib inline
plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)
plt.xlim(11,15)
plt.ylim(0,30)
# plt.show()
```

#### <span id="c62" style="color:Green"> 6.2 Adding axis labels  </span>

```python
d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]


%matplotlib inline
plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)
plt.xlabel('Date',size=12,fontweight='bold')
plt.ylabel('Temperature (°C)',size=12,fontweight='semibold')
plt.show()
```

#### <span id="c63" style="color:Green"> 6.3 Adding grid  </span>

```python
d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]


%matplotlib inline
plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)
plt.grid(True,linewidth=0.5)
plt.xlabel('Date',size=12,fontweight='bold')
plt.ylabel('Temperature (°C)',size=12,fontweight='semibold')
plt.show()
```

#### <span id="c64" style="color:Green"> 6.4 Adding title  </span>

```python
d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]


%matplotlib inline
plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)
plt.grid(True,linewidth=0.5)
plt.xlabel('Date',size=12,fontweight='bold')
plt.ylabel('Temperature (°C)',size=12,fontweight='semibold')
plt.title('Daily Temps of 3 cities in the second week of December',size=16,fontweight='bold')
plt.show()
```

#### <span id="c65" style="color:Green"> 6.5 Adding legend  </span>

To match a data series on plots with their labels, such as by their line styles and marker styles, we add plt.legend()

Inorder to adjust the location we pass loc settings. Possible loc settings are as follows:-

<ul>
	<li>'best':0(onlyimplementedforaxeslegends)</li>
	<li>'upperright':1</li>
	<li>'upperleft':2</li>
	<li>'lowerleft':3</li>
	<li>'right':5(thesameas'centerright';forback-compatibility)</li>
	<li>'centerleft':6</li>
	<li>'centerright':7</li>
	<li>'lowercenter':8</li>
	<li>'uppercenter':9</li>
	<li>'center':10</li>
</ul>

```python
d = [11,12,13,14,15,16,17]
t0 = [15.3,12.6,12.7,13.2,12.3,11.4,12.8]
t1 = [26.1,26.2,24.3,25.1,26.7,27.8,26.9]
t2 = [22.3,20.6,19.8,21.6,21.3,19.4,21.4]


%matplotlib inline
plt.plot(d,t0)
plt.plot(d,t1)
plt.plot(d,t2)
plt.grid(True,linewidth=0.5)
plt.legend(('Tokyo','Hawaii','Hong Kong'))
plt.xlabel('Date',size=12,fontweight='bold')
plt.ylabel('Temperature (°C)',size=12,fontweight='semibold')
plt.title('Daily Temps of 3 cities in the second week of December',size=16,fontweight='bold')
plt.show()
```

### <span id="c7" style="color:Green"> 7. Saving plots to a file  </span>
Inorder to save a figure we put plt.savefig(outputpath) at the end of plotting commands.If you want to save the figure as a file as well as display it on the notebook output, you can call both plt.savefig() and plt.show().

If no file extension is mentioned then PNG format is obtained as output. This supports a number of image formats, including PNG, JPG, PDF and PostScript.

```python
y = np.linspace(1,2000)
x = 1.0 / np.sin(y)
plt.plot(x,y,'green')
plt.xlim(-20,20)
plt.ylim(1000,2400)
plt.savefig('123',dpi=120) 
# dpi means 'Dots per inch'(8 * 12)inches i.e (8 * 300)*(12 * 300)=2400 * 3600 pixels in image
plt.show()
```

<h4> 1.1 Color Cycle </h4>

A color cycle is a list of colors used to control the color of a series automatically, such as each data series in multiline plots.These colors are designed to show good contrast between distinct categories.Each color is named 'C0' to 'C9', and can be called in manually by specifying a color in the preset color cycle.

```python
i = 3
plt.plot([i]*6)
plt.xlim(0,10)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4,4))
for color in colors:
#     color = np.random.choice(colors)
    plt.plot([np.random.randint(10)]*5,c=color,label=color) 
plt.legend()
plt.xlim(0,6)
plt.show()
```

#### 1.2 Single-letter abbreviations for basic colors
There are several common colors with built-in. Some are listed below:-

* 'b':Blue
* 'g':Green
* 'r':Red
* 'c':Cyan
* 'm':Magenta
* 'y':Yellow
* 'k':Black
* 'w':White

```python
colors=['b','g','r','c','m','y','k','w']
for color in colors:
    plt.plot([1]*5,c=color,label=color)
```

```python
colors=['b','g','r','c','m','y','k','w']
for color in colors:
#     print(color)
    plt.plot(np.random.randint(5)*5,c=color,label=color)
plt.legend()
plt.show()
```

#### 1.3 RGB or RGBA color code
A color can be specified as a tuple of three or four float numbers between 0 to 1 such as (0.1,0.1,0.2) or (0.2,0.2,0.3,0.8).The first three colors define how red, green and blue should be mixed into the desired color output. The optional fourth value is the alpha value to control the transparency level.

### <span id="a2" style="color:Green"> 2 More native Matplotlib plot types   </span> 
Besides the most basic scatter plot and line plots, Matplotlib provides a versatile collection of plot types to serve different data visualization purposes.

A successful visualization must communicate the message well. Inorder to achieve this goal, we need to have good understanding of our data as well as advantages and limitations of plot type in illustrating differnt relationships in data.

In choosing the right plot type to display, we have the following considerations:

Number of variables
Distribution of data
Relationship between data series

#### 2.1 Histogram

Histograms are useful in surveying the distribution of data. For example, we can plot data on a histogram when we want to see some age groups distributed in a population, light exposure in a photograph, or the amount of precipitation in each month in a city.

In Matplotlib, we call the plt.hist() function with a linear array. Matplotlib will automatically group them into 'bins' and plot out the frequencies for each bin in bars

```python
np.random.binomial(10,0.9,10)
```

```python
help(plt.hist2d)
```

```python
np.random.seed(42)
x = np.random.binomial(10,0.9,size=10000)
y = np.random.binomial(10,0.9,size=10000)
plt.hist2d(x,y,20)
plt.show()
```
