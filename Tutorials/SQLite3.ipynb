#### ORDER BY desc
* ORDER BY orders the result by based on a column

```python
Query = """
    select number,invoice_total,invoice_date,due_date,payment_date
       from invoices 
       order by invoice_total desc;
"""
execquery(Query,con).head()
```

#### ORDER BY
* By default ORDER BY orders the data in ascending order

```python
Query = """
    select number,invoice_total,invoice_date,due_date,payment_date
       from invoices 
       order by invoice_total;
"""
execquery(Query,con).head()
```

#### WHERE
* WHERE clause is used to extract data based on some condition

```python
Query = """
    select *
       from invoices 
       where client_id=5;
"""
execquery(Query,con).head()
```

#### Class Task:
* Retrieve all invoice observations where the payment is not done

```python
Query = """
    select *
       from invoices 
       where payment_total=0;
"""
execquery(Query,con).head()
```

### Comparison operators
* <,>,<=,>=,<> are other comparison operators which can be used with WHERE clause

```python
Query = """
    select *
       from invoices 
       where invoice_total>150;
"""
execquery(Query,con).head()
```

```python
Query = """
    select *
       from invoices 
       where client_id <> 5;
"""
execquery(Query,con).head()
```

### Logical operators
* AND,OR,NOT,BETWEEN,IN

#### AND

```python
Query = """
    select *
       from invoices 
       where invoice_total>150 and payment_total=0;
"""
execquery(Query,con).head()
```

#### Between

```python
Query = """ 
    SELECT * FROM EMPLOYEES;
"""
execquery(Query,con).head()
```

```python
Query = """ 
    SELECT * 
      FROM EMPLOYEES
      where salary between 70000 and 100000
    ;
"""
execquery(Query,con).head()
```

#### Not

```python
Query = """ 
    SELECT * 
      FROM EMPLOYEES
      where salary not between 70000 and 100000
    ;
"""
execquery(Query,con).head()
```

#### IN 

```python
Query = """ 
    SELECT * 
      FROM EMPLOYEES
      where office_id in (2,3)
    ;
"""
execquery(Query,con).head()
```

### Like & Wild Cards

### Null Values

```python
Query = """ 
    SELECT *
     FROM invoices 
     WHERE payment_date is null;
"""
execquery(Query,con)
```

####  Class Task 
* Extracting rows where the payment has been made

```python
Query = """ 
    SELECT *
     FROM invoices 
     WHERE payment_date is not null;
"""
execquery(Query,con).head()
```

### Ifnull

```python
Query = """ 
    SELECT ifnull(payment_date,'0000-00-00') as payment_date_v1
     FROM invoices;
"""
execquery(Query,con).head()
```

#### Class Task: 
* Extract "invoices" table with a new column "Date_Diff" which denotes number of days since invoice and payment. For cases where payment is not done consider current date as payment date

```python
Query = """
    SELECT *,ifnull(strftime('%J',payment_date) - strftime('%J',invoice_date),date('now')) as Date_Diff
    FROM invoices;    
"""
execquery(Query,con).head()
```
